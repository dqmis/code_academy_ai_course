```python
import pandas as pd
```

```python
df = pd.read_csv("./data/Car details v3.csv")
```

```python
df.groupby("owner").count()
```

```python
owner_map = {
    "Test Drive Car": 0,
    "First Owner": 1,
    "Second Owner": 2,
    "Third Owner": 3,
    "Fourth & Above Owner": 4
}
```

```python
df["owner_encoded"] = df.owner.apply(lambda x: owner_map[x])
```

```python
df.groupby("owner_encoded").count()
```

```python
df["brand"] = df.name.apply(lambda x: x.split()[0])
```
