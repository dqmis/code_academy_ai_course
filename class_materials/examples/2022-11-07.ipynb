```python
def div(a: int, b: int) -> int:
    return int(a / b)

def div(a: int, b: int) -> int:
    try:
        return int(a / b)
    except ZeroDivisionError:
        return 0

def div(a: int, b: int) -> int:
    if b != 0:
        return int(a / b)
    return 0
```
